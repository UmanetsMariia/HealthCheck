```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
```

### DF overview

```python
df = pd.read_csv("diabetes.csv")
```

```python
df.head()
```

```python
df.drop(columns=['location'], inplace=True)
```

```python
def describe(df):
    summary_data = []
    
    for column in df.columns:
        variable = column
        dtype = df[column].dtype
        count = len(df[column])
        unique = len(df[column].unique())
        missing = df[column].isna().sum()
        
        if dtype in ['float64', 'int64']:
            min_val, max_val = df[column].min(), df[column].max()
        else:
            min_val, max_val = 'Str', 'Str'
        
        summary_data.append([variable, dtype, count, unique, missing, min_val, max_val])
    
    output_columns = ['variable', 'dtype', 'count', 'unique', 'missing value', 'Min', 'Max']
    output = pd.DataFrame(summary_data, columns=output_columns)
    
    return output

main = describe(df)
main
```

### Handling with missing values

bp.1s = First diastolic blood pressure <br>
bp.1d = First systolic blood pressure <br>
chol = Cholesterol<br>
stab.glu = Stabilized Glucose<br>
hdl = High Density Lipoprotein Cholesterol<br>
hip = Hybrid insulin peptides<br>
time.ppn = Partial parenteral nutrition

glyhb - target var (>= 6.5 - labelled as positive diabetes)

First step - deleting rows with missing values for chol, hdl, ratio, weight, waist,heigt, hip, time.ppn - amont of them is small - won'r affect the reoresentativity

```python
df2 = df.dropna(subset=['chol', 'hdl', 'ratio', 'weight', 'waist', 'height', 'hip', 'time.ppn', 'glyhb', 'bp.1s', 'bp.1d'])
```

```python
main = describe(df2)
main
```

```python
df2["outcome"] = df2.glyhb >= 6.5
df2["outcome"] = df2["outcome"].astype(int)
df2
```

let's see how bp.1s, bp.1d and bp.2s correlate with target var

```python
df2.drop(columns=['glyhb', 'bp.2s', 'bp.2d'], inplace=True)
df2['gender_encoded'] = df2['gender'].apply(lambda x: 1 if x.lower() == 'female' else 0)
df2.drop(columns=['gender'], inplace=True)
```

```python
data = df2.copy()
# Split the data into two subsets: one with missing "frame" values and one without
data_with_frame = data[~data['frame'].isnull()]
data_missing_frame = data[data['frame'].isnull()]

# Select your feature columns and the target column
X = data_with_frame.drop(['frame'], axis=1)
y = data_with_frame['frame']

# Train a RandomForestClassifier
model = RandomForestClassifier()
model.fit(X, y)
# Impute missing values using the trained model
missing_frame_values = model.predict(data_missing_frame.drop(['frame'], axis=1))
data_missing_frame['frame'] = missing_frame_values

# Combine the imputed data with the original data
imputed_data = pd.concat([data_with_frame, data_missing_frame])
address_dummy = pd.get_dummies(imputed_data['frame'])
imputed_data = imputed_data.merge(address_dummy, left_index = True, right_index = True)
imputed_data.drop(columns = 'frame', inplace = True)

frame_name = imputed_data.columns[-3::]
imputed_data[frame_name] = imputed_data[frame_name].astype('float64')

outcome_column = imputed_data['outcome']
imputed_data = imputed_data.drop(columns=['outcome'])
imputed_data['outcome'] = outcome_column
imputed_data.head()
```

```python
imputed_data.loc[:, 'bmi'] = 703 * (imputed_data['weight'] / imputed_data['height']**2)
imputed_data.drop(columns=['weight', 'height'], inplace=True)
```
